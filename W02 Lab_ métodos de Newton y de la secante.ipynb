# Método de Newton

## **Ejercicio 1:**
Considere la función
$$f(x)=e^{x^2}+(2x-1)^3.$$

a) Implemente y grafique la función $f$.

b) Implemente $f'$ y $f''$ en funciones de Python.  

c) Utilice el método de Newton para hallar el mínimo.
- Pruebe 3 valores diferentes valores para el punto inicial e indique si el algoritmo converge al valor que minimiza la función en cada caso.
- Considere como condición de parada 100 iteraciones.

d) Implemente el algoritmo con condición de parada de modo que la precisión sea $\epsilon=10^{-5}$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return np.exp(x**2)+(2*x-1)**3

x = np.arange(-2, 2, 0.01)    # arreglo de posibles valores de x
y=np.array([f(t) for t in x])  #
plt.plot(x, y)
plt.show()
```

```python
from scipy.optimize import minimize_scalar
```

```python
minimize_scalar(f, method="golden", bracket=[0.5,1.0])
```

```python
def f_prima(x):
  return 2*x*np.exp(x**2) + 6*(2*x-1)**2
```

```python
def f_segunda_derivada(x):
  return 2*np.exp(x**2) + 4*x**2*np.exp(x**2)+24*(2*x-1)
```

```python
x=-2
valores=[]
for i in range(10):
  x=x-f_prima(x)/f_segunda_derivada(x)
  valores.append(f(x))
print(x)
```

```python
plt.plot(valores)
```

```python
x=0
tolerancia=10**-5
distancia=tolerancia+1
iter=0
while (distancia>tolerancia):
  x1=x-f_prima(x)/f_segunda_derivada(x)
  distancia=np.abs(x1-x)
  x=x1
  iter+=1
print(x)
print(iter)
```


# Método de la secante

## **Ejercicio 2:**
Considere la función
$$f(x)=\dfrac{1}{2}x^2-\text{sen} x.$$

a) Implemente y grafique la función $f$.

b) Implemente el Método de la Secante en una función y utilicela para hallar el mínimo de la función con una precisión de $\epsilon=10^{-5}$.

```python
def f(x):
  return 1/2*x**2-np.sin(x)

x = np.arange(-3, 5, 0.01)    # arreglo de posibles valores de x
y=np.array([f(t) for t in x])  #
plt.plot(x, y)
plt.show()
```

```python
def f_prima(x):
  return x-np.cos(x)
```

```python
x_anterior=1
x=0
tolerancia=10**-5
distancia=tolerancia+1
iter=0
while (distancia>tolerancia):
  x_nuevo=x-f_prima(x)/(f_prima(x)-f_prima(x_anterior))*(x-x_anterior)
  distancia=np.abs(x_nuevo-x)
  x_anterior=x
  x=x_nuevo
  iter+=1
print(x)
print(iter)
```
