# Dirección de máximo decrecimiento

Considera la función

$$f(x, y)=10y^4+(1-x)^2.$$

## **Ejercicio 1:**

a) Implementa la función $f$ y su gradiente.

b) Halla la dirección de decrecimiento máximo en el punto $(3, 0.5)$.

```python
import numpy as np
```

```python
def f(punto):
  x=punto[0]
  y=punto[1]
  return 10*y**4+(1-x)**2
```

```python
def grad_f(punto):
  x=punto[0]
  y=punto[1]
  return np.array([ -2*(1-x), 40*y**3 ])
```

Dirección de decrecimiento máximo

```python
-grad_f([3, 0.5])
```

## **Ejercicio 3:**
Si $v$ es la dirección de máximo decrecimiento, defina la función
$$g(\alpha)=f((3, 0.5)+\alpha v)$$
y grafíquela para $\alpha>0$. ¿En qué valor de $\alpha$ alcanza su mínimo?

```python
def g(a):
  return f(np.array([3, 0.5])+a*np.array([-4, -5]))
```

```python
import matplotlib.pyplot as plt
x=np.arange(0, 0.5, 0.01)
y=[g(k) for k in x]
plt.plot(x, y)
plt.show()
```

```python
from scipy.optimize import minimize_scalar
```

```python
minimize_scalar(g, method="brent").x
```
